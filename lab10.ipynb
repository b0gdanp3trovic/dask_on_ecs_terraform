# Dask delayed() function

This is a toy example with two very simple fulnctions that sleep for a while to simulate work.

```python
from time import sleep

def inc(x):
 sleep(1)
 return x + 1

def add(x, y):
 sleep(1)
 return x + y
```

```python
%%time
x = inc(5)
y = inc(10)
z = add(x,y)
```

The two increment calls can be executed in parallel as they are independent of eachother.

```python
from dask import delayed
x = delayed(inc)(1)
y = delayed(inc)(2)
z = delayed(add)(x, y)
```

```python
%%time
z.compute()
```

```python
z.visualize()
```

```python
def square(x):
 sleep(1)
 return x*x

def double(x):
 sleep(1)
 return x*2

def add(x, y):
 sleep(1)
 return x + y
```

```python
%%time
output = []
for i in range(6):
 a = square(i)
 b = double(i)
 c = add(a, b)
 output.append(c)

total= sum(output)
```

```python
%%time
output = []
for i in range(6):
 a = delayed(square)(i)
 b = delayed(double)(i)
 c = delayed(add)(a, b)
 output.append(c)

total= delayed(sum)(output)
total.compute()
total.visualize()
```

Unlike pandas.read_csv which reads in the entire file before inferring datatypes, dask.dataframe.read_csv only reads in a sample from the beginning of the file (or first file if using a glob). These inferred datatypes are then enforced when reading all partitions.

The inferred data types can be incorrect. When this happens you have a few options:

* Specify dtypes directly using the dtype keyword. This is the recommended solution, as it's the least error prone (better to be explicit than implicit) and also the most performant.
* Increase the size of the sample keyword (in bytes)
* Use assume_missing to make dask assume that columns inferred to be int (which don't allow missing values) are actually floats (which do allow missing values). In our particular case this doesn't apply.

```python
%time rides.trip_distance.max().compute()
```

```python
%time rides.trip_distance.max().visualize()
```

```python
mean_fare = rides.fare_amount.mean()
std_fare= rides.fare_amount.std()
```

```python
%%time
mean_fare = mean_fare.compute()
std_fare = std_fare.compute()
```

```python
%%time
import dask
mean_fare, std_fare = dask.compute(mean_fare, std_fare)
```

This way the two results are merged when calling dask.compute, allowing shared operations to only be done once instead of twice.

## persist()

If we run compute() we are asking Dask to take all the computations and adjust to the data that we have queued up, and run them, and bring it the the surfacace right here. If it was distributed we covert it into a local object. 
It basically transforms a Dask dataframe into a Pandas dataframe.

If we use .persist() , we are asking Dask to take all the computations and adjustments to the data that we have queued up, and run them, but then
the object is going to remain distributed and will live on the cluster (a LocalCluster if you are on one machine), not on the Jupyter instance or other local environment. It basically keeps a distributed Dask dataframe.

```python
rides = rides[rides.trip_distance > 10]
```

```python
rides = rides.assign(fares = rides.fare_amount + rides.tip_amount) 
```

```python
%time rides.fares.max().compute()
```

```python
rides = client.persist(rides)
```

```python
%time rides.fares.max().compute()
```

Naturally, persisting every intermediate along the way is a bad idea, because this will tend to fill up all available RAM and make the whole system slow (or break!). The ideal persist point is often at the end of a set of data cleaning steps, when the data is in a form which will get queried often.
